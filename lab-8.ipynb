```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
covid_data = np.array([
    [1500, 2000, 1800, 1200, 900],  # Day 1
    [1600, 2100, 1900, 1300, 950],  # Day 2
    [1700, 2200, 2000, 1400, 1000], # Day 3
    [1650, 2150, 1950, 1350, 980],  # Day 4
    [1750, 2250, 2050, 1450, 1020], # Day 5
    [1800, 2300, 2100, 1500, 1050], # Day 6
    [1900, 2400, 2200, 1600, 1100], # Day 7
])

# Countries
countries = ["Country_A", "Country_B", "Country_C", "Country_D", "Country_E"]

# 1. **Basic Statistics**
# Total cases reported in each country
total_cases = np.sum(covid_data, axis=0)
print("Total cases for each country:", total_cases)

# Country with the highest total cases
max_country_index = np.argmax(total_cases)
max_country = countries[max_country_index]
print("Country with the highest total cases:", max_country)

# 2. **Daily Analysis**
# Average number of cases per day across all countries
daily_avg_cases = np.mean(covid_data, axis=1)
print("Average daily cases across all countries:", daily_avg_cases)

# Day with the highest total cases across all countries
total_daily_cases = np.sum(covid_data, axis=1)
highest_day_index = np.argmax(total_daily_cases) + 1
print(f"Day with the highest total cases is Day {highest_day_index}.")

# 3. **Trends**
# Percentage increase/decrease for each country from Day 1 to Day 7
percentage_change = ((covid_data[-1] - covid_data[0]) / covid_data[0]) * 100
print("Percentage change for each country:", percentage_change)

# Country with the highest percentage increase
max_increase_index = np.argmax(percentage_change)
max_increase_country = countries[max_increase_index]
print("Country with the highest percentage increase:", max_increase_country)

# 4. **Data Transformation**
# Normalizing data for each country
min_vals = np.min(covid_data, axis=0)
max_vals = np.max(covid_data, axis=0)
normalized_data = (covid_data - min_vals) / (max_vals - min_vals)
print("Normalized dataset:\n", normalized_data)

# 5. **Visualization**
plt.figure(figsize=(12, 6))

# Line chart for daily cases of each country
for i in range(len(countries)):
    plt.plot(range(1, 8), covid_data[:, i], label=countries[i])

# Highlight the day with the highest total cases
plt.axvline(x=highest_day_index, color='red', linestyle='--', label=f"Highest Cases (Day {highest_day_index})")

plt.title("COVID-19 Daily Cases for Each Country")
plt.xlabel("Day")
plt.ylabel("Daily Cases")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
covid_data = np.array([
    [1500, 2000, 1800, 1200, 900],  # Day 1 
    [1600, 2100, 1900, 1300, 950],  # Day 2 
    [1700, 2200, 2000, 1400, 1000],  # Day 3 
    [1650, 2150, 1950, 1350, 980],  # Day 4 
    [1750, 2250, 2050, 1450, 1020],  # Day 5 
    [1800, 2300, 2100, 1500, 1050],  # Day 6 
    [1900, 2400, 2200, 1600, 1100],  # Day 7 
])

countries = ["Country_A", "Country_B", "Country_C", "Country_D", "Country_E"]

# Basic Statistics
total_cases = covid_data.sum(axis=0)
country_highest_cases = countries[np.argmax(total_cases)]

# Daily Analysis
daily_avg_cases = covid_data.mean(axis=1)
day_highest_cases = np.argmax(covid_data.sum(axis=1)) + 1

# Trends
percentage_change = ((covid_data[-1] - covid_data[0]) / covid_data[0]) * 100
country_highest_increase = countries[np.argmax(percentage_change)]

# Data Normalization
normalized_data = (covid_data - covid_data.min(axis=0)) / (covid_data.max(axis=0) - covid_data.min(axis=0))

# Results
print("Total Cases per Country:", dict(zip(countries, total_cases)))
print("Country with Highest Total Cases:", country_highest_cases)
print("Daily Average Cases:", daily_avg_cases)
print("Day with Highest Total Cases:", day_highest_cases)
print("Percentage Change per Country:", dict(zip(countries, percentage_change)))
print("Country with Highest Percentage Increase:", country_highest_increase)
print("Normalized Dataset:\n", normalized_data)

# Visualization
plt.figure(figsize=(12, 6))

# Plot daily cases for each country
for i, country in enumerate(countries):
    plt.plot(range(1, 8), covid_data[:, i], marker='o', label=country)

# Highlight the day with the highest total cases
highest_day_total_cases = covid_data.sum(axis=1).max()
highest_day = covid_data.sum(axis=1).argmax() + 1
plt.axvline(x=highest_day, color='red', linestyle='--', label=f"Day {highest_day} Highest Cases")

# Add title and labels
plt.title("Daily COVID-19 Cases per Country")
plt.xlabel("Day")
plt.ylabel("Number of Cases")
plt.legend()
plt.grid()
plt.tight_layout()

# Show plot
plt.show()
```
