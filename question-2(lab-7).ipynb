```python
import random

def separate_primes_and_composites(numbers):
    primes = []
    composites = []

    for num in numbers:
        if num < 2:
            continue  # Skip numbers less than 2 (not prime or composite)
        
        # Check if number is prime or composite
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(num)
        else:
            composites.append(num)
    
    return primes, composites

# Generate random numbers as per user-defined K and N
N = int(input("Enter the upper limit (N) for random numbers: "))
K = int(input("Enter the number of random numbers to generate (K): "))

# Ensure K is at least 10
if K < 10:
    print("K should be at least 10. Setting K to 10.")
    K = 10

# Generate K random numbers within the range [0, N]
random_numbers = [random.randint(0, N) for _ in range(K)]
print("Generated list of random numbers:", random_numbers)

# Separate the primes and composites
primes, composites = separate_primes_and_composites(random_numbers)

print("Prime numbers:", primes)
print("Composite numbers:", composites)
```

```python
import random
import math

def separate_and_calculate(numbers):
    primes = []
    primes_squared = []
    composites = []
    composites_sqrt = []

    for num in numbers:
        if num < 2:
            continue  # Skip numbers less than 2 (not prime or composite)
        
        # Check if number is prime or composite
        is_prime = True
        for i in range(2, int(math.sqrt(num)) + 1):
            if num % i == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(num)
            primes_squared.append(num ** 2)
        else:
            composites.append(num)
            composites_sqrt.append(math.sqrt(num))
    
    return primes, primes_squared, composites, composites_sqrt

# Generate random numbers as per user-defined K and N
N = int(input("Enter the upper limit (N) for random numbers: "))
K = int(input("Enter the number of random numbers to generate (K): "))

# Ensure K is at least 10
if K < 10:
    print("K should be at least 10. Setting K to 10.")
    K = 10

# Generate K random numbers within the range [0, N]
random_numbers = [random.randint(0, N) for _ in range(K)]
print("Generated list of random numbers:", random_numbers)

# Separate primes and composites, and calculate squares and square roots
primes, primes_squared, composites, composites_sqrt = separate_and_calculate(random_numbers)

print("Prime numbers:", primes)
print("Square of prime numbers:", primes_squared)
print("Composite numbers:", composites)
print("Square root of composite numbers:", composites_sqrt)
```

```python
import random
import math
import matplotlib.pyplot as plt

def separate_and_calculate(numbers):
    primes = []
    primes_squared = []
    composites = []
    composites_sqrt = []

    for num in numbers:
        if num < 2:
            continue  # Skip numbers less than 2 (not prime or composite)
        
        # Check if number is prime or composite
        is_prime = True
        for i in range(2, int(math.sqrt(num)) + 1):
            if num % i == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(num)
            primes_squared.append(num ** 2)
        else:
            composites.append(num)
            composites_sqrt.append(math.sqrt(num))
    
    return primes, primes_squared, composites, composites_sqrt

# Generate random numbers as per user-defined K and N
N = int(input("Enter the upper limit (N) for random numbers: "))
K = int(input("Enter the number of random numbers to generate (K): "))

# Ensure K is at least 10
if K < 10:
    print("K should be at least 10. Setting K to 10.")
    K = 10

# Generate K random numbers within the range [0, N]
random_numbers = [random.randint(0, N) for _ in range(K)]
print("Generated list of random numbers:", random_numbers)

# Separate primes and composites, and calculate squares and square roots
primes, primes_squared, composites, composites_sqrt = separate_and_calculate(random_numbers)

# Plotting
plt.figure(figsize=(12, 5))

# Subplot for Prime numbers vs. their squares
plt.subplot(1, 2, 1)
plt.scatter(primes, primes_squared, color='blue', label="Prime Numbers")
plt.xlabel("Prime Numbers")
plt.ylabel("Square of Prime Numbers")
plt.title("Prime Numbers vs. Their Squares")
plt.legend()

# Subplot for Composite numbers vs. their square roots
plt.subplot(1, 2, 2)
plt.scatter(composites, composites_sqrt, color='red', label="Composite Numbers")
plt.xlabel("Composite Numbers")
plt.ylabel("Square Root of Composite Numbers")
plt.title("Composite Numbers vs. Their Square Roots")
plt.legend()

plt.tight_layout()
plt.show()
```
